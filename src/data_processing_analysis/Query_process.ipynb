### 读取并保存数据

```python
import pandas as pd
import json

# 读取第一个 JSON 文件
with open(r'../../data/raw/all_steam_data1.json', 'r', encoding='utf-16') as file1:
    data1 = json.load(file1)

# 读取第二个 JSON 文件
with open(r'../../data/raw/all_steam_data2.json', 'r', encoding='utf-16') as file2:
    data2 = json.load(file2)

# 将 JSON 数据转换为 pandas DataFrame
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# 拼接两个 DataFrame
df = pd.concat([df1, df2], ignore_index=True)
df = df.drop(1506)
df 
```

将数据整理成一列便于后续提取数据

```python
# 使用 pd.concat 方法将所有列的数据合并成一列
combined_df = pd.concat([df[col] for col in df.columns], ignore_index=True)

# 将合并后的数据转换为 DataFrame
combined_df = pd.DataFrame(combined_df, columns=['steam_data'])

# 显示合并后的 DataFrame
df = combined_df
df = df.drop(7529)
```

### 查询数据

#### 定义查询函数:player_info类

```python
'''定义提取player_info中的数据'''
def extract_player_info(cell):
    try:
        key = list(cell.keys())[0]
        players = cell[key]['player_info']['response']['players'][0]
        
        return pd.Series({
            'id': key,
            'personaname': players.get('personaname', None),
            'profileurl': players.get('profileurl', None),
            'avatar': players.get('avatar', None),
            'timecreated': players.get('timecreated', None),
            'loccountrycode': players.get('loccountrycode', None)
        })
    except Exception as e:
        # print(f"Error: {e}")  # 打印错误信息以便调试
        return pd.Series({
            'id': None,
            'personaname': None,
            'profileurl': None,
            'avatar': None,
            'timecreated': None,
            'loccountrycode': None
        })
```

```python
'''提取所需数据'''
extract_df = df['steam_data'].apply(extract_player_info)
```

```python
'''将数据补充至新df'''
df_for_save_data =pd.concat([extract_df],ignore_index=True)
df_for_save_data
```

#### 定义查询函数:player_games类

```python
'''定义提取player_games中的数据'''
def extract_player_games(cell):
    key = list(cell.keys())[0]
    try:       
        response = cell[key]['player_games']['response']
        games = response.get('games', [])

        # 按游玩时长排序并获取前五个游戏
        top_games = sorted(games, key=lambda x: x['playtime_forever'], reverse=True)[:5]
        
        # 初始化默认值
        result = {
            'id':key,
            'game_count': response.get('game_count', None),
            'game_1_name': None, 'game_1_playtime': None,'game_1_id': None,
            'game_2_name': None, 'game_2_playtime': None,'game_2_id': None,
            'game_3_name': None, 'game_3_playtime': None,'game_3_id': None,
            'game_4_name': None, 'game_4_playtime': None,'game_4_id': None,
            'game_5_name': None, 'game_5_playtime': None,'game_5_id': None,
        }

        # 填充前五个游戏的信息
        for i, game in enumerate(top_games):
            result[f'game_{i+1}_name'] = game.get('name', None)
            result[f'game_{i+1}_playtime'] = game.get('playtime_forever', None)
            result[f'game_{i+1}id'] = game.get('playtime_forever', None)
        
        return pd.Series(result)
    except Exception as e:
        # print(f"Error: {e}")  # 打印错误信息以便调试
        return pd.Series({
            'id':key,
            'game_count': None,
            'game_1_name': None, 'game_1_playtime': None,
            'game_2_name': None, 'game_2_playtime': None,
            'game_3_name': None, 'game_3_playtime': None,
            'game_4_name': None, 'game_4_playtime': None,
            'game_5_name': None, 'game_5_playtime': None,
        })


```

```python
extract_df = df['steam_data'].apply(extract_player_games)
extract_df
```
